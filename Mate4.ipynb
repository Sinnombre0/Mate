```python
from sympy import symbols, cos, sin, exp, pi
from sympy import plot_parametric
t = symbols('t')
escala = 0.125
def x(t):
  coseno = cos(t)
  return  coseno ** 3
def y(t):
  seno = sin(t)
  return seno ** 3
def z(t):
  return (3 * cos(t)- cos(3 * t))
def w(t):
  return (3 * sin(t) + sin(3 * t))

plot_parametric((x(t), y(t), (t, -6*pi, 6*pi)),(escala * z(t), escala * w(t),(t, -6*pi, 6*pi)));
```

```python
from sympy import symbols, cos, sin, exp, pi, tan, sec
from sympy import plot_parametric
t = symbols('t')
a = 2
def x(t):
  return a * (1 + cos(t)) * cos(t)
def y(t):
  return a * (1 + cos(t)) * sin(t)
def z(t):
  return 2 * a + 3 * a * cos(t) * (1 - cos(t))
def w(t):
  return 3 * a * sin(t) * (1 - cos(t))

plot_parametric((x(t), y(t), (t, -6*pi, 6*pi)),(z(t), w(t),(t, -6*pi, 6*pi)));
```

```python
from sympy import symbols, cos, sin, exp, pi, tan, sec
from sympy import plot_parametric
t = symbols('t')
a = 100
def x(t):
  return a * (t - sin(t))
def y(t):
  return a * (1 - cos(t))
def z(t):
  return a * (t + sin(t))
def w(t):
  return a * (3 + cos(t))

plot_parametric((x(t), y(t), (t, -6*pi, 6*pi)),(z(t), w(t),(t, -6*pi, 6*pi)));
```

```python
from sympy import symbols, cos, sin, exp, pi, tan, sec
from sympy import plot_parametric
t = symbols('t')
escala = 0.333
escala2 = 0.111
def x(t):
  return 2 * cos(t) + cos(2 * t)
def y(t):
  return 2 * sin(t) - sin(2 * t)
def z(t):
  return 2 * cos(t) - cos(2 * t)
def w(t):
  return 2 * sin(t) + sin(2 * t)

plot_parametric((escala * x(t), escala * y(t), (t, -6*pi, 6*pi)),(escala2 * z(t),escala2 * w(t),(t, -6*pi, 6*pi)));
```

```python
from sympy import symbols, cos, sin, exp, pi, tan, sec
from sympy import plot_parametric
t = symbols('t')
h = 0.25
def x(t):
  return exp(h * t) * cos(t)
def y(t):
  return exp(h * t) * sin(t)
def z(t):
  return (exp(h * t) * sin(t))/h
def w(t):
  return -(exp(h * t) * cos(t))/h

plot_parametric((x(t), y(t), (t, -6*pi, 6*pi)),(z(t), w(t),(t, -6*pi, 6*pi)));
```
